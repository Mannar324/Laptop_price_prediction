## Extract Some Features

```python
def extract_features(text,pattern):
    matches=re.findall(pattern,text,re.IGNORECASE)
    return matches
```

```python
#get inch
def search_pattern(text):
    match = regex.search(text)
    return match.group() if match else np.nan
regex = re.compile("\d+(\.\d+)?\s*-*Inch")


# Apply the search function to the 'Laptop_name' column
matches = data2['Laptop_name'].apply(search_pattern)
data2['Inch']=matches
```

```python
#get Cpu
pattern=r'(?:Intel\s+)?Core\s+i\d'
core=data2['Laptop_name'].apply(lambda x:x.split(','))
core=core.apply(lambda x: str(x[1]) if len(x) > 1 else 'None')
Cores=core.apply(lambda x:extract_features(x,pattern))
Cores=Cores.apply(lambda x:''.join(x) if x else np.nan)
data2['Core']=Cores
```

```python
#get cpu
regex = re.compile("(?:Intel\s+)?Core\s+i\d")
matches = data2['Laptop_name'].apply(search_pattern)
data2['Core']=matches
```
